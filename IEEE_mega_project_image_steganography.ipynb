Importing PIL(Python Image Library) to extract pixel values from image and modify it.

```python
import numpy as np
from PIL import Image
```

Function to convert the data entered by the user into 8-bit binary :

```python
def conv_bin(data): 
    newdata = [] 
    for i in data: 
        newdata.append(format(ord(i), '08b')) 
    return newdata
```

Function to modify the pixels of the image to encode the data :

```python
def modify_pixel(pixel, data): 
    newdata= conv_bin(data) 
    image_data = iter(pixel) 

    for i in range(len(newdata)):
        pixel = [value for value in next(image_data)[:3] + next(image_data)[:3] + next(image_data)[:3]]
        for j in range(0, 8): 
            if (newdata[i][j]=='0') and (pixel[j]% 2 != 0): 
                    pixel[j] -= 1
            elif (newdata[i][j] == '1') and (pixel[j] % 2 == 0): 
                pixel[j] -= 1
        
        if (i == len(newdata) - 1): 
            if (pixel[-1] % 2 == 0): 
                pixel[-1] -= 1
        else: 
            if (pixel[-1] % 2 != 0): 
                pixel[-1] -= 1

        pixel = tuple(pixel) 
        yield pixel[0:3] 
        yield pixel[3:6] 
        yield pixel[6:9]
```

Function to change the values of the pixels in the new image :

```python
def encode_img(newimage, data): 
    w = newimage.width
    (x, y) = (0, 0) 
    for pixel in modify_pixel(newimage.getdata(), data):
        newimage.putpixel((x, y), pixel) 
        if (x == w - 1): 
            x = 0
            y += 1
        else: 
            x += 1
```

Function to encode and save the new image :

```python
def encode(): 
    original = input("Enter image name(with extension): ") 
    image = Image.open(original, 'r') 
    
    data = input("Enter data to be encoded : ") 

    newimage = image.copy() 
    encode_img(newimage, data)  
    newimage.save("modified.png") 
```

Function to decode the modified image :

```python
def decode(): 
    img = input("Enter image name(with extension) :") 
    image = Image.open(img, 'r') 

    data = '' 
    imgdata = iter(image.getdata()) 

    while (1): 
        pixels = [value for value in next(imgdata)[:3] + next(imgdata)[:3] + next(imgdata)[:3]]
        binstr = '' 

        for i in pixels[:8]: 
            if (i % 2 == 0): 
                binstr += '0'
            else: 
                binstr += '1'

        data += chr(int(binstr, 2))
        if (pixels[-1] % 2 != 0): 
            return data 
```

```python
 a = int(input("Choose 1 to Encode or 2 to Decode\n")) 
if (a == 1): 
    encode() 

elif (a == 2): 
    print("Decoded sentence is", decode())
```
